```python
import pandas as pd
import numpy as np
```

```python
Tabla_Tiempos=pd.read_excel('Matriz_Tiempos.xlsx',sheet_name='Hoja1')
Tabla_Tiempos = Tabla_Tiempos.set_index('Unidad')
```

```python
Tabla_Piezas=pd.read_excel('Matriz_Tiempos.xlsx',sheet_name='Hoja2')
Tabla_Piezas = Tabla_Piezas.set_index('Pieza')
Tabla_Piezas['A abastecer'] = Tabla_Piezas['Packing']*Tabla_Piezas['Abast. simultáneo']
Tabla_Piezas
```

```python
Tabla_Repo = pd.DataFrame(0, index=Tabla_Tiempos.index, columns=Tabla_Piezas.index)
for u in Tabla_Tiempos.index:
  for pieza in Tabla_Piezas.index:
    if u==1:
      Tabla_Tiempos[pieza] = 0
      Tabla_Tiempos.loc[u,pieza] = Tabla_Piezas.loc[pieza,'Stock inicial']
      if Tabla_Tiempos.loc[u,pieza] < Tabla_Piezas.loc[pieza,'Uso']:
        Tabla_Tiempos.loc[u,pieza] += Tabla_Piezas.loc[pieza,'A abastecer']
        Tabla_Repo.loc[u,pieza] = 'SIN STOCK'
    elif u=='FIN':
      Tabla_Tiempos.loc[u,pieza] = Tabla_Tiempos.loc[len(Tabla_Tiempos)-1,pieza] - Tabla_Piezas.loc[pieza,'Uso']
      if Tabla_Tiempos.loc[u,pieza] < Tabla_Piezas.loc[pieza,'Uso']:
        Tabla_Tiempos.loc[u,pieza] += Tabla_Piezas.loc[pieza,'A abastecer']
        Tabla_Repo.loc[u,pieza] = 'SIN STOCK'
    else:
      Tabla_Tiempos.loc[u,pieza] = Tabla_Tiempos.loc[u-1,pieza] - Tabla_Piezas.loc[pieza,'Uso']
      if Tabla_Tiempos.loc[u,pieza] < Tabla_Piezas.loc[pieza,'Uso']:
        Tabla_Tiempos.loc[u,pieza] += Tabla_Piezas.loc[pieza,'A abastecer']
        Tabla_Repo.loc[u,pieza] = 'SIN STOCK'
Tabla_Tiempos.head(5)
```

```python
Tabla_Repo.head(5)
```

```python
for pieza in Tabla_Repo.columns:
  Tabla_Repo[pieza] = Tabla_Repo[pieza].replace('SIN STOCK', pieza)
Tabla_Repo.head(5)
```

```python
Tabla_Resumen = pd.DataFrame(columns=['Sin stock a las:', 'Pieza a reponer:', 'A abastecer:', 'Responsable:'])
Tabla_Repo['Hora'] = Tabla_Tiempos['Hora']
i=0
for u in Tabla_Repo.index:
  for pieza in Tabla_Repo.columns[:-1]:
    if Tabla_Repo.loc[u,pieza] != 0:
      Tabla_Resumen.loc[i,'Pieza a reponer:'] = Tabla_Repo.loc[u,pieza]
      Tabla_Resumen.loc[i,'Sin stock a las:'] = Tabla_Repo.loc[u,'Hora']
      Tabla_Resumen.loc[i,'A abastecer:'] = Tabla_Piezas.loc[pieza,'A abastecer']
      Tabla_Resumen.loc[i,'Responsable:'] = Tabla_Piezas.loc[pieza,'Operario']
      i += 1
Tabla_Resumen.to_excel('Tabla_Resumen.xlsx', index=False)
Tabla_Resumen
```
